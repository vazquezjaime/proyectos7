```python
import pandas as pd
import plotly.express as px

# Leer los datos
car_data = pd.read_csv('vehicles_us.csv')

# Opcional: eliminar valores nulos del odmetro
car_data = car_data.dropna(subset=['odometer'])

# Crear histograma
fig = px.histogram(
    car_data,
    x='odometer',
    nbins=50,
    title='Distribucin del kilometraje'
)

# Mostrar grfico
fig.show()

import pandas as pd
import plotly.express as px

car_data = pd.read_csv('vehicles_us.csv')

# Limpieza bsica
car_data = car_data.dropna(subset=['odometer', 'price'])

fig = px.scatter(
    car_data,
    x='odometer',
    y='price',
    title='Precio vs kilometraje',
    opacity=0.5
)

fig.show()

```
